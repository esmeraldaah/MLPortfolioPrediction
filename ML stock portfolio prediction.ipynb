```python
import pandas as pd
import datetime as dt
from sklearn import preprocessing
import numpy as np
```

```python
df = pd.read_csv("df500_6m_adjclose.csv")
df.head()
```

```python
# Veryfing that all stocks have full historical data in a 4 year period
count_stocks = df.describe()
count_stocks.iloc[0][count_stocks.iloc[0,:] != 1009] 
```

```python
# Extract only monthly data
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.loc[(df['Date'].dt.day==7) , :] 
df.head()
```

```python
# Get stock returns
df.iloc[:,1:] = (df.iloc[:,1:]/df.iloc[:,1:].shift(1))-1 # (p(t+1)/p(t))-1
df = df.iloc[1:,:]
df.head()
```

```python
# Normalize stocks
scaler = preprocessing.MinMaxScaler()
names = df.columns
d = scaler.fit_transform(df.iloc[:,1:])
df.iloc[:,1:] = d
df.head()
```

```python
# Create correlation and select low correlated stocks
rets_corr = df.corr().abs().unstack().sort_values()
todelete=[]
rets_corr
for i in np.arange(len(rets_corr)):
    if (float(rets_corr[i]) < 0.002):
        todelete.append(i)
rets_corr = pd.DataFrame(rets_corr).reset_index()
rets_corr = rets_corr.loc[0:max(todelete)]
rets_corr
```

```python
# Filter data with uncorrelated stocks
stocks_lowc = list(set(rets_corr.iloc[:,0]))
len(stocks_lowc)
stocks_lowc.insert(0, 'Date')
stocks_lowc
df = df[stocks_lowc]
df.head()
```

```python
# Create lags for each stock
X_lags = []
Y_lags = []
column_names=[]
for i in range(len(df.T)):
    if i == 0:
        continue
    else:
        for n in range(13):
            if n == 0:
                X_lags.append(df.iloc[:,i])
            else:    
                Y_lags.append(df.iloc[:,i].shift(-n))
                column_names.append(df.columns[i] + "_lag_" + str(n))
Y = pd.DataFrame(Y_lags).T
Y.columns = column_names
Y = Y.reset_index(drop=True)

X = pd.DataFrame(X_lags).T
X.columns = stocks_lowc[1:]
X = X.reset_index(drop=True)
Y.tail(20)
```
